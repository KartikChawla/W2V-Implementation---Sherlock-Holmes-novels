```python
file = open("A Study in Scarlet.txt")
nov_sis = file.read()
file.close()
file = open("The Hound of Baskervilles.txt")
nov_hob = file.read()
file.close()
file = open('The Sign of the Four.txt')
nov_sof = file.read()
file.close()
file = open('The Valley of Fear.txt')
nov_vof = file.read()
file.close()
```

Data Preparation 

```python
full_novels = nov_sis + nov_hob + nov_sof + nov_vof
all_novels = [nov_sis, nov_hob, nov_sof, nov_vof]
```

```python
puncts = [string.punctuation]
wtoke_sh= []
for file in all_novels:
    tokens = word_tokenize(full_novels)
    wtoke_sh.append(tokens)
    
print(len(wtoke_sh))
```

```python
tokes = [word for word in wtoke_sh[1] if not word in stop_words]
print (wtoke_sh[1][1:100])
print('\n')
print (tokes[1:100])
```

```python
filtered = []
for i in range (1,4):
    filtered.append([word for word in wtoke_sh[i] if not word in stop_words])
    filtered.append([word for word in wtoke_sh[i] if not word in puncts])
    filtered.append([word for word in wtoke_sh[i] if len(word) > 1])
```

```python
word_total = len(filtered)
print(word_total)
print(filtered[1:100])
```

```python
shmodel = gensim.models.Word2Vec(filtered, min_count =1, window =10)
shmodel.train(filtered, total_words=shmodel.corpus_count, epochs = shmodel.epochs)
```
